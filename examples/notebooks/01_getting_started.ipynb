# Getting Started with pyMyriad

This notebook introduces the basic workflow: create data → build analysis tree → run analysis → view results.

```python
import pandas as pd
import numpy as np
from pyMyriad import AnalysisTree
```

## Create Sample Data

Create a simple dataset with demographics and measurements.

```python
np.random.seed(42)
df = pd.DataFrame({
    "Gender": np.random.choice(["M", "F"], 200),
    "Country": np.random.choice(["US", "UK", "Canada"], 200),
    "Age": np.random.randint(18, 70, 200),
    "Income": np.random.normal(50000, 15000, 200),
})

df.head()
```

## Build an Analysis Tree

Define your analysis structure by specifying:
- **Splits**: Group data by variables (e.g., Gender, Country)
- **Analysis**: Compute statistics for each group

### String Expressions vs Lambda Functions

You can specify analyses in two ways:
1. **String expressions**: `"np.mean(df.Income)"` - evaluated in the context of the data
2. **Lambda functions**: `lambda df: np.mean(df.Income)` - standard Python functions

Both approaches work identically. Use `df` to refer to the dataframe at each node.

```python
# Build analysis tree: split by Gender, then analyze Income
atree = AnalysisTree()\
    .split_by("df.Gender")\
    .analyze_by(
        mean_income=lambda df: np.mean(df.Income),
        std_income=lambda df: np.std(df.Income),
        count=lambda df: len(df)
    )

# Print the analysis plan
print(atree)
```

## Run the Analysis

Apply the analysis tree to your data to get results.

```python
# Run analysis on the dataframe
result = atree.run(df)

# Print the results
print(result)
```

## View Results as a Table

Convert results to a pandas DataFrame for easy viewing.

```python
from pyMyriad import simple_table

simple_table(result)
```
