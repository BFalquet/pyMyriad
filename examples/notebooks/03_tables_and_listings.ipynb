# Tables and Listings

This notebook demonstrates how to format analysis results as tables using simple_table(), gt_table(), and cascade_table().

```python
import pandas as pd
import numpy as np
from pyMyriad import AnalysisTree, simple_table, gt_table, cascade_table
```

```python
# Create sample data and run analysis
np.random.seed(42)
df = pd.DataFrame(
    {
        "Gender": np.random.choice(["M", "F"], 200),
        "Country": np.random.choice(["US", "UK", "Canada"], 200),
        "Age": np.random.randint(18, 70, 200),
        "Income": np.random.normal(50000, 15000, 200),
    }
)

atree = (
    AnalysisTree()
    .split_by("df.Gender")
    .split_by("df.Country")
    .analyze_by(
        mean_income=lambda df: np.mean(df.Income),
        std_income=lambda df: np.std(df.Income),
        count=lambda df: len(df),
    )
)

result = atree.run(df)
```

## simple_table() - Basic DataFrame Output

Convert results to a pandas DataFrame with hierarchical path columns.

```python
simple_table(result)
```

## Pivoting with `by` Parameter

Pivot results to show split levels as columns instead of rows.

```python
# Pivot by Country - each country becomes a column
simple_table(result, by="df.Country")
```

```python
# Pivot by Gender - each gender becomes a column
simple_table(result, by="df.Gender")
```

## gt_table() - Enhanced Formatting

Create publication-ready tables with great_tables styling.

```python
# Basic gt_table with hierarchical display
gt_table(result)
```

```python
# gt_table with pivoting
gt_table(result, by="df.Country")
```

## cascade_table() - Cascading Layout

Display results in a cascading format for easier reading.

```python
cascade_table(result).head(10)
```

## Cascade layout in gt table

Use `cascade=True` in `gt_table()` to get a cascading layout. 

```python
# Create analysis with summarize_by
atree2 = (
    AnalysisTree()
    .split_by("df.Gender")
    .summarize_by(overall_mean=lambda df: np.mean(df.Income))
    .split_by("df.Country")
    .analyze_by(mean_income=lambda df: np.mean(df.Income), count=lambda df: len(df))
)

result2 = atree2.run(df)

# Include summary statistics in output
gt_table(result2, cascade=True)
```

## Pivot Statistics

Create wide-format tables with statistics as separate columns.

```python
# Each statistic becomes a separate column
simple_table(result, pivot_statistics=True)
```

## Combining Options

Use multiple parameters together for complex table layouts.

```python
# Pivot by Country with statistics as columns
simple_table(result, by="df.Country", pivot_statistics=True)
```
